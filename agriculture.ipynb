```python
df=df.drop_duplicates()
```

```python
print (df.shape)
```

```python
df = df.dropna()
```

```python
df=df.fillna(df.mean(numeric_only=True))
```

```python
df=df.fillna("unknown")
```

```python
df.columns=df.columns.str.strip()
for col in df.select_dtypes(include="object"):
    df[col]=df[col].str.strip()
```
